```python
data = xr.open_dataset(r"/home/nowwhat/swt_proj/era5_19120612.nc", group=None).squeeze()#Squeeze notwendig für Interpolation (streicht time als dimension)
data = data.metpy.parse_cf()#Wird u.a. benötigt, um cross_section() nutzen zu können
```

```python
def my_cross_section(data_var_arr, start, end):
    data_var_arr_csec = []
    for var in data_var_arr:
        #print(var)
        data_var_arr_csec.append(cross_section(var, start, end))
        #print(data[var])
    #list = [cross_section(arr[varr], start, end) for var in arr]
    
    return data_var_arr_csec
```

```python
def interpolate(data_var_arr, plevs, csec = False):
    interpolated_vars = []
    for var in data_var_arr:
        arrlike = log_interpolate_1d(plevs, data_var_arr[0],  var)#pressure is first
        
        #Back to DataArray:
        if csec:
            tmp = xr.DataArray(data=arrlike,
                               dims=['plevs', 'index'], 
                               coords={'plevs': ('plevs', plevs), 
                                       'index': ('index', var['index'])}, 
                               attrs=var.attrs)
        else:
            tmp = xr.DataArray(data=arrlike,
                               dims=['plevs', 'lat', 'lon'],
                               coords={'plevs': ('plevs', plevs),
                                       'lat': ('lat', var['lat']), 
                                       'lon': ('lon', var['lon'])},
                                attrs=var.attrs)
        interpolated_vars.append(tmp)
    return interpolated_vars
```

```python
def plot_csec(data_arrays, plevs):
    fig = plt.figure(figsize=(15, 10))
    ax = plt.axes()
    for var in data_arrays:
        if var.long_name=='Temperature':
            cf = ax.contourf([1*i for i in range(0,100)], plevs,data_arrays[1], 20, cmap='YlGnBu')
            cb = fig.colorbar(cf, orientation='horizontal', extend='max', aspect=65, shrink=0.5, pad=0.1, extendrect='True')
            cb.set_label(data.TEMP.units, size='x-large')

    ax.set_yscale('symlog')
    ax.set_ylim(plevs.max(), plevs.min()) 
    plt.show()
```

```python
def plot_on_level(data_arrays, level):
    fig = plt.figure(figsize=(15, 10))
    ax = plt.axes(projection=ccrs.PlateCarree())
    #ax._hold = True  # Work-around for CartoPy 0.16/Matplotlib 3.0.0 incompatibility --> Still needed?
    cmap = plt.cm.Pastel2
    for var in data_arrays:
        #print(var.long_name)
        if var.long_name=='Temperature':
            cs = plt.contourf(var.lon,
                      var.lat,
                      var[level, : :],
                      20,
                      cmap='YlGnBu',
                      transform=ccrs.PlateCarree())
            cbar = plt.colorbar(cs, orientation='horizontal', fraction=.05)
        #elif var.long_name == 'U component of wind':
         #   csu = ax.contour(var.lon, var.lat, var[level, : :], 15, colors='k', linewidths=1.5)
          #  csu.clabel(csu.levels[1::2], fontsize=8, colors='k', inline=1, inline_spacing=8, fmt='%i', rightside_up=True, use_clabeltext=True)

    ax.coastlines()
    ax.gridlines()
    plt.show()
#plot_on_level([data.TEMP], 2)
```

```python
#Creates plot (and necessary data manipulation) based on the users input
#PLevs as param
def do_it_csec(sel_vars, start=(0,0), end=(10,10)):
    plevs = [1000., 800., 900.,700.,600.,500.,400.,300.,200.,100.,90.,80.,70.,60.,50.,40.,30.,20.,10.]*units.hPa
    plevs[::-1].sort()

    plotvariables = list(sel_vars.value)
    #plotvariables.remove('PRESS')
    plotvariables.insert(0, 'PRESS')#Always have Pressure as first list element
    data_arrs = [data[var] for var in plotvariables]   

    #csec_vars = my_cross_section(data_arrs, start, end)
    csec_vars = [cross_section(var, start, end) for var in data_arrs]
    interpolated_vars = interpolate(csec_vars, plevs, csec=True)
    plot_csec(interpolated_vars, plevs)
```

```python
#area: [left, bottom, right, top] -- Dict may be better

def do_it_onlevel(sel_vars, area, level=4):
    plevs = [1000., 800., 900.,700.,600.,500.,400.,300.,200.,100.,90.,80.,70.,60.,50.,40.,30.,20.,10.]*units.hPa
    plevs[::-1].sort()
    #plotvariables = sel_vars
    plotvariables = list(sel_vars.value)
    #plotvariables.remove('PRESS')
    plotvariables.insert(0, 'PRESS')#Pressure necessary for later interpolation
    data_arrs = [data[var] for var in plotvariables]  

    data_arrs = sliceit(data_arrs, area)
    #isel vertical level first to boost interpolation time?
    interpolated_vars = interpolate(data_arrs, plevs)
    plot_on_level(interpolated_vars, level) 
#do_it_onlevel(['TEMP'], [10, 10, 20, 20])
```

```python
#Show selected area on map
def area_on_map(left, bottom, width, height):
    plt.figure(1, figsize=(16, 9))
    ax = plt.axes(projection=ccrs.PlateCarree())
    ax.stock_img()
    
    if left >=180:
        left -= 360#Notice: value changed only for plot (local variable)
    rect = mplpatches.Rectangle([left, bottom], width, height)
    ax.add_patch(rect)
    plt.show()

#SHOW 2 POINTS ON MAP (FOR CROSS SECTION)
def points_on_map(lon1, lat1, lon2, lat2):
    plt.figure(1, figsize=(16, 9))
    ax = plt.axes(projection=ccrs.PlateCarree())
    ax.stock_img()
    plt.plot(lon1, lat1, '+-r' , transform=ccrs.PlateCarree(), ms=40)
    plt.plot(lon2, lat2, '+-m' , transform=ccrs.PlateCarree(), ms=40)
    #plt.plot(lon1, lat1, '+-r' , transform=ccrs.PlateCarree(), ms=40)
    #plt.plot(lon2, lat2, '+-m' , transform=ccrs.PlateCarree(), ms=40)
    #plt.plot([sel_lon.value, sel_lon2.value], [sel_lat.value, sel_lat2.value], color='gray', linestyle='--', transform=ccrs.PlateCarree())
    plt.show()
    
```
